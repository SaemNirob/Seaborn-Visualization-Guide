## Cat Plot

#### Cat Plot In Seaborn
 - Categorical scatterplots with catplot

 - stripplot() – with kind = "strip"

 - swarmplot() – with kind = "swarm"

#### Categorical distribution plots with catplot

 - boxplot() – with kind = "box"

 - violinplot() – with kind = "violin"

 - boxenplot() – with kind = "boxen"

#### Categorical estimate plots with catplot

 - pointplot() – with kind = "point"

 - barplot() – with kind = "bar"

 - countplot() – with kind = "count"

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(x = 'tip' , y = 'size', data = data)

plt.show()
```

```python
data2 = sns.load_dataset('tips').head(20)
```

```python
sns.catplot(x = 'tip' , y = 'size', data = data2)

plt.show()
```

```python
sns.catplot(x = 'tip' , y = 'size', data = data2, hue = 'sex')

plt.show()
```

```python
sns.catplot(x = 'tip' , y = 'size', data = data2, hue = 'sex', palette = 'Oranges')

plt.show()
```

```python
sns.catplot(x = 'tip' , y = 'size', data = data2, hue = 'sex', palette = 'Oranges',
           height = 5 )

plt.show()
```

```python
sns.catplot(x = 'day' , y = 'size', data = data, hue = 'sex', palette = 'Oranges',
           kind = 'bar')

plt.show()
```

```python
sns.catplot(x = 'day' , y = 'size', data = data, hue = 'sex', palette = 'Oranges',
           kind = 'point')

plt.show()
```

```python
sns.catplot(x = 'day' , y = 'total_bill', data = data, palette = 'Oranges',
           kind = 'box')

plt.show()
```

```python
sns.catplot(x = 'day' , y = 'total_bill', data = data, palette = 'Oranges',
           kind = 'boxen')

plt.show()
```
