## Violin Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = data)

plt.show()
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = data, linewidth = 3)

plt.show()
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = data, linewidth = 1.5, palette = 'Dark2')

plt.show()
```

```python
sns.violinplot(x = 'day', y ='total_bill', data = data, linewidth = 1.5, palette = 'Dark2',
              saturation = 0.4)

plt.show()
```

```python
sns.violinplot(x = 'time', y = 'total_bill', data = data, palette = 'Dark2', order = ['Dinner', 'Lunch'])
```

```python
sns.violinplot(x = 'time', y = 'total_bill', data = data, palette = 'Dark2', order = ['Dinner', 'Lunch'], inner = 'point')

plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data, palette = 'bright', hue ='sex', split = True)

plt.show()
```
