# Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var  = [1, 2, 3, 4, 5, 6, 7]
var2 = [2, 3, 4, 1, 5, 6, 7]
```

```python
df = pd.DataFrame({'var' : var, 'var2' : var2})
df
```

```python
sns.lineplot(x = 'var', y = 'var2', data = df)

plt.show()
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data)

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex')

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', size = 10)

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', style = 'sex')

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', style = 'sex',
             palette = 'Accent')

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', style = 'sex',  
             palette = 'Accent', markers = ['o', '>'])

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', style = 'sex', 
             palette = 'Accent', markers = ['o', '>'], dashes = False)

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', style = 'sex', 
             palette = 'Accent', markers = ['o', '>'], dashes = False, legend = False)

plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue ='sex', style = 'sex', 
             palette = 'Accent', markers = ['o', '>'], dashes = False, legend = False)

plt.grid()

plt.title('Python')

plt.show()
```
