## Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data)

plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'g')

plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'g',
            hue = 'time')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'g',
            hue = 'sex')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'g',
            hue = 'sex', order = ['Fri', 'Sat', 'Sun', 'Thur'])

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'r',
           order = ['Fri', 'Sat', 'Sun', 'Thur'], showmeans = True)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'r',
           order = ['Fri', 'Sat', 'Sun', 'Thur'], showmeans = True,
           meanprops = {'marker' : '+', 'markeredgecolor' : 'w'})

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'r',
           order = ['Fri', 'Sat', 'Sun', 'Thur'], showmeans = True,
           meanprops = {'marker' : '+', 'markeredgecolor' : 'w'},
           linewidth = 3)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = data, color = 'r',
           order = ['Fri', 'Sat', 'Sun', 'Thur'], showmeans = True,
           meanprops = {'marker' : '+', 'markeredgecolor' : 'w'},
           linewidth = 3, palette = 'plasma')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(y = 'day', x = 'total_bill', data = data, color = 'r',
           order = ['Fri', 'Sat', 'Sun', 'Thur'], showmeans = True,
           meanprops = {'marker' : '+', 'markeredgecolor' : 'w'},
           linewidth = 3, palette = 'plasma')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = data['total_bill'], palette = 'plasma')

plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(y = data['total_bill'], palette = 'plasma')

plt.show()
```
