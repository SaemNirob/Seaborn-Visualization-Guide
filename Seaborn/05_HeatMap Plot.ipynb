# HeatMap Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
array = np.linspace(1, 10, 20).reshape(4,5)
array
```

```python
sns.heatmap(array)

plt.show()
```

```python
arr = np.linspace(1, 10, 10).reshape(2, 5)
arr
```

```python
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr', annot =True)

plt.show()
```

```python
arr1 = np.array([['a0', 'a1', 'a2', 'a3', 'a4'],
                 ['b0', 'b1', 'b2', 'b3', 'b4']])
arr1
```

```python
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's')

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x)

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2)

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2,
           linecolor = 'black')

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2,
           linecolor = 'black', cbar = False)

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2,
           linecolor = 'black', cbar = False, xticklabels = False)

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2,
           linecolor = 'black', cbar = False, xticklabels = False, yticklabels = False)

plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
v = sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2,
           linecolor = 'black', cbar = False, xticklabels = False, yticklabels = False)
v.set(xlabel = 'Python', ylabel = 'Data')
plt.show()
```

```python
x = {'fontsize' : 15, 'color' : 'b'}
v = sns.heatmap(arr, vmin = 0, vmax = 10, cmap = 'PuOr',
            annot = arr1, fmt = 's', annot_kws = x, linewidths = 2,
           linecolor = 'black', cbar = False, xticklabels = False, yticklabels = False)

v.set(xlabel = 'Python', ylabel = 'Data')
sns.set(font_scale = 3)

plt.show()
```
