## Facet Grid

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
fg = sns.FacetGrid(df)

fg.map(plt.scatter, 'total_bill', 'tip')

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'sex')

fg.map(plt.scatter, 'total_bill', 'tip')

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'sex', hue = 'day')

fg.map(plt.scatter, 'total_bill', 'tip')

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'sex', hue = 'day')

fg.map(plt.scatter, 'total_bill', 'tip').add_legend()

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'day', hue = 'sex')

fg.map(plt.scatter, 'total_bill', 'tip').add_legend()

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'day', hue = 'sex')

fg.map(plt.bar, 'total_bill', 'tip').add_legend()

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'sex', hue = 'day')

fg.map(plt.bar, 'total_bill', 'tip').add_legend()

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'sex', hue = 'day', palette = 'summer')

fg.map(plt.bar, 'total_bill', 'tip').add_legend()

plt.show()
```

```python
fg = sns.FacetGrid(df, col = 'sex', hue = 'day', palette = 'summer')

fg.map(plt.bar, 'total_bill', 'tip', edgecolor = 'red').add_legend()

plt.show()
```
