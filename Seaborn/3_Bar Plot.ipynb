# Bar Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data)

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex')

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'])

plt.show()
```

```python
order  = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
            order = order)

plt.show()
```

```python
order  = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
            order = order, ci = 99)

plt.show()
```

```python
order  = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = data, color = 'g')

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], palette = 'icefire')

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], saturation = 0.3)

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], saturation = 0.3, errcolor = 'b')

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], saturation = 0.3, errcolor = 'b', errwidth = 5)

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], capsize = 0.4)

plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], capsize = 0.4, alpha = 0.6)

plt.show()
```

```python
sns.set(style = 'darkgrid')
sns.barplot(x = 'island', y = 'bill_length_mm', data = data , hue = 'sex',
           hue_order = ['Female', 'Male'], capsize = 0.4)

plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", hue="sex",data=tips)
```

```python
tips[tips['day']=='Thur']['total_bill'].sum()
```

```python
import numpy as np
#sns.barplot(x='day', y='total_bill', data=tips, estimator=np.std)
sns.barplot(x="day", y="total_bill", data=tips, estimator=np.sum)
```

```python
datalabel = sns.barplot(x="day", y="total_bill", data=tips, estimator=np.sum, ci=None)

for i in datalabel.containers:
    datalabel.bar_label(i)
```

```python
datalabel=sns.barplot(x="day", y="total_bill", data=tips, ci=None, estimator=np.std)
for i in datalabel.containers:
    datalabel.bar_label(i)
```
