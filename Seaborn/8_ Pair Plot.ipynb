## Pair Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.pairplot(data)

plt.show()
```

```python
sns.pairplot(data, hue = 'sex')

plt.show()
```

```python
sns.pairplot(data, vars = ['total_bill', 'tip'], hue = 'sex')

plt.show()
```

```python
sns.pairplot(data, vars = ['total_bill', 'tip'], hue = 'sex', hue_order = ['Female', 'Male'])

plt.show()
```

```python
sns.pairplot(data, vars = ['total_bill', 'tip'], hue = 'sex', 
             hue_order = ['Female', 'Male'], palette = 'BuGn')

plt.show()
```

```python
sns.pairplot(data, hue = 'sex', hue_order = ['Female', 'Male'], 
             kind = 'reg') # 'scatter', 'kde', 'hist', 'reg'

plt.show()
```

```python
sns.pairplot(data, hue = 'sex', hue_order = ['Female', 'Male'], 
             kind = 'reg', diag_kind = 'hist') # 'scatter', 'kde', 'hist', 'reg'

plt.show()
```

```python
sns.pairplot(data, hue = 'sex', markers =['*', '>'])
             

plt.show()
```
