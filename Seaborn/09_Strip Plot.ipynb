## Strip Plot 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data)

plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex')

plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8, edgecolor = 'blue')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8, jitter = 0.3)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8, size = 8)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8, size = 5, marker = '<')

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
m = {'Male' : '*', 'Female' : 'o'}
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8, size = 5, marker = m)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', palette = 'rocket_r',
             linewidth = 0.8, size = 8, alpha = 0.5)

plt.legend(loc = 0, fontsize = 8)
plt.show()
```

```python
sns.stripplot(x = data['total_bill'])

plt.show()
```
