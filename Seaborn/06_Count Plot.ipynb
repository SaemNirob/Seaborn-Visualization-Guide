## Count Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.countplot(x = 'sex', data = data)

plt.show()
```

```python
sns.countplot(x = 'sex', data = data, hue = 'smoker')

plt.show()
```

```python
sns.countplot(x = 'sex', data = data, hue = 'smoker', color = 'r')

plt.show()
```

```python
sns.countplot(x = 'sex', data = data, hue = 'smoker', palette = 'bright')

plt.show()
```

```python
sns.countplot(x = 'sex', data = data, hue = 'smoker', palette = 'bright', saturation = 0.3)

plt.show()
```

```python
datalabel = sns.countplot(x = 'sex', data = data, hue = 'smoker', palette = 'bright')

for i in datalabel.containers:
    datalabel.bar_label(i)
    
plt.show()
```
