# Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data)

plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data, hue = 'sex')

plt.show()
```

```python
data2 = sns.load_dataset('penguins').head(20)
data2.shape
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, hue = 'sex')

plt.legend(loc = 4)
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex')

plt.legend(loc = 4)
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex', size = 'sex')

plt.legend(loc = 4)
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex', size = 'sex', sizes = (100, 60))
               

plt.legend(loc = 4)
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex', size = 'sex', sizes = (100, 60),
               palette = 'gist_ncar')
               

plt.legend(loc = 4)
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex', size = 'sex', sizes = (100, 60),
               palette = 'gist_ncar', alpha = 0.5)
               

plt.legend(loc = 4)
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex', size = 'sex', sizes = (100, 60),
               palette = 'gist_ncar', alpha = 0.5, markers = ('o', '*'))
               

plt.legend(loc = 4)
plt.show()
```

```python
m = {'Male' : 'o', 'Female' : '*'}
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = data2, 
                hue = 'sex', style = 'sex', size = 'sex', sizes = (100, 60),
               palette = 'gist_ncar', alpha = 0.5, markers = m)
               

plt.legend(loc = 4)
plt.show()
```
